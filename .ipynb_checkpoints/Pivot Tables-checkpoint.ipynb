```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('./survey_results_public.csv')
```

```python
df2=df.select_dtypes(include='object')
df3=df.select_dtypes(include=['float','int'])
df3['Professional'] =df['Professional']
df3['Country']=df['Country']
```

```python
df3.columns
```

```python
df3.pivot_table(index=['Professional'])
## non numeric columns are skipped in the colujns section
```

```python
df3.pivot_table(index=['Professional','Country'])
## Default mean of columns
```

```python
df3.pivot_table(index=['Professional','Country'],columns=['CareerSatisfaction'])
## CareerSatisfaction is treated like categories #Rest columns are averaged
```

```python
df3.pivot_table(index=['Professional','Country'],columns=['CareerSatisfaction'],values=['HoursPerWeek'])
## CareerSatisfaction is treated like categories # select only columns specified in values
```

```python
df3.pivot_table(index=['Professional','Country'],columns=['CareerSatisfaction'],values=['HoursPerWeek'],aggfunc=np.mean)
## CareerSatisfaction is treated like categories # select only columns specified in values
```

```python
df_sum=df3.pivot_table(index=['Professional','Country'],columns=['CareerSatisfaction'],values=['HoursPerWeek'],aggfunc=np.sum)
## CareerSatisfaction is treated like categories # select only columns specified in values # instead of mean calculate sum
df_sum
#df_sum.loc['Professional developer']
```

```python
df_sum.loc['Professional developer']
```

```python
print(type(df_sum.index))
print("Multi INdex is a tuple of columns")
df_sum.index

```

```python
## You can use first field in the tuple to get the second field
df_sum.loc['Professional developer'].index
```

```python
print("Columns is also multi-index tuple")
df_sum.columns
```

```python
df4=df3.pivot_table(index=['Professional','Country'],columns=['CareerSatisfaction'],values=['HoursPerWeek'],aggfunc=np.mean)
df4
```

```python
df4.loc['Professional developer']
```

```python
df4.loc['Professional developer'].loc['Australia']
```

```python
# set of aggregation functions for all columns
df5=df3.pivot_table(index=['Professional','Country'],columns=['CareerSatisfaction'],values=['HoursPerWeek','Salary'],aggfunc=[np.sum,np.mean])
df5
```

```python
df5.loc['Professional developer'].loc['United States']
```

```python

df5.columns
```

```python
df5.loc['Professional developer'].loc['United States'][( 'sum', 'HoursPerWeek',  9.0)]
## [(aggregation,value,pivoted_column)]
```

```python
#one agregation function for each column
df6=df3.pivot_table(index=['Professional','Country'],columns=['CareerSatisfaction'],aggfunc={'HoursPerWeek':np.mean,'Salary':np.mean})
df6
```

```python
df6.loc['Professional developer'].loc['United States']
## I took np.mean for all columns
```

```python
df6.loc['Professional developer'].loc['United States'][('HoursPerWeek',1.0)]
# tuple is the column
```

```python
df6.loc['Professional developer'].loc['United States'][('Salary',9.0)]
```

```python
from sklearn.datasets import load_iris
iris=load_iris()

iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df
```

```python
## Each row to 
iris.target_names
```

```python
iris_df.columns
```

```python
## Each row corresponds to one data point
## Lets arrange data point vertically like Key-Value style
# row index is not preserved
iris_df.melt(value_vars=iris_df.columns)
```

```python
iris_df.melt(value_vars=iris_df.columns).sort_index()
```

```python
iris_reset=iris_df.reset_index()
# creates a new index column
iris_reset
```

```python
# Specify row identifier in id_vars
iris_reset.melt(id_vars=['index'],value_vars=iris_df.columns).sort_values(['index'])
```

```python
iris_reset.melt(id_vars=['index'],value_vars=iris_df.columns,value_name='New',var_name='Attributes').sort_values(['index'])
```

```python
iris_melt=iris_reset.melt(id_vars=['index']).sort_values(['index'])
iris_melt
```

```python
iris_melt.pivot_table(index=['index'],columns=['variable'],values=['value'])
```

```python
iris_melt.pivot_table(index=['index'],columns=['variable'],values=['value']).columns
```
