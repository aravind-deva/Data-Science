```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.plot(kind='bar')
plt.show()
```

```python
df.iloc[0,1]=-1.5
```

```python
df.style.bar(subset=[df.columns[1]],align='mid',color=['red', 'green'])
```

```python
df[df.columns[0]].hist(bins=10)
```

```python
df[df.columns[1]].value_counts().plot(kind='bar')
```
