# Apply vs Transform
- There are improper explanations out there on the web. 

## Let us Understand two types of functions

- A function which takes a list of values and returns list of values of the same length
    - In pandas world, a series(row/column) is a list of values and return is a series(new rows/columns)
    - In pandas world, a data frame is a list of series
- A function which takes a list of values and returns a scalar 
        - In pandas world , a scalar can be something like sum()/mean() of all values
        - In pandas world, a data frame is a list of series

## APPLY
- APPLY on either rows/columns => series accepts both type of functions
        1. You can use **result_type='broadcast'** if dimensions are reduced in second type of function
- APPLY if a dataframe is passed in (group by returns groups of DF) accepts both type of functions
        2. There is **NO** result_type='broadcast' if dimensions are reduced in second type of function
        3. Refer GroupBy.apply
        
## TRANSFORM
- TRANSFORM accepts only **first type** of function.
        - TRANSFORM accepts second type of function only when you pass in a DataFrame, it automatically broadcasts results to each group 
- TRANSFORM  if a dataframe is passed in (group by returns groups of DF) ,for each group,the function is applied on a series(rows/columns) and the result is broad casted if the dimensions are reduced
- EFFECTIVELY transform is applicable for input type - Series. **transform on DataFrameGroupBy** object does the magic of 
broadcasting 
   - **DataFrameGroupBy.transform**

```python
def add(x):
    print('----')
    print(f"Passed in -- {type(x)}")
    print(x)
    print(f"Passed in length {len(x)} return length {len(x+10)}")
    print('----')
    return x+10
```

```python
df.apply(add) #default axis=0 => columns
```

```python
df.transform(add) #default axis=0 => columns
```

```python
def my_mean(x):
    print('----')
    print(f"Passed in -- {type(x)}")
    print(x)
    print(f"passed in length {len(x)} return length =1 scalar {x.mean().shape}")
    print(f"return value {x.mean()}")
    print('----')
    return x.mean()
```

```python
df.apply(my_mean) #default axis=0 => columns

#For every column only one scalar(mean) is returned
```

```python
df.apply(my_mean,result_type='broadcast')
```

```python
df.transform(my_mean) #default axis=0 => columns
print("Transform does not accept change in length of the given axis.here default axis =0" )
```

```python
df[df.columns[1]].unique()
df[df.columns[1]].nunique()
df[df.columns[1]].value_counts()
```

```python
df.groupby(df.columns[1]).apply(add).sort_values(df.columns[1],ascending=True)

## Applies for each group
```

```python
apply_group=df.groupby(df.columns[1]).apply(my_mean)
print(apply_group.shape)
print(apply_group)
```

```python
df.groupby(df.columns[1]).apply(my_mean,result_type='broadcast')
```

```python
#df.groupby(df.columns[1])
print(df.groupby(df.columns[1]).transform(add))


```

```python
t_groupby=df.groupby(df.columns[1]).transform(my_mean)
print(t_groupby)
print(t_groupby.loc[[62,68,119]])
```

```python
groups=df.groupby(df.columns[1])

#print(groups.get_group(2.0).transform(add))
for name,group in groups:
    print(name)
    print()
    print(type(group))
    #print(group.head(2))
    print()
    print(group.transform(add))
```
