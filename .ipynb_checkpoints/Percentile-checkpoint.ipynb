```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(4)
p = np.linspace(0, 100, 6001)
ax = plt.gca()
lines = [
('linear', None),
('higher', '--'),
('lower', '--'),
('nearest', '-.'),
('midpoint', '-.'),
]
for interpolation, style in lines:
    ax.plot(p, np.percentile(a, p, interpolation=interpolation),label=interpolation, linestyle=style)

ax.set(title='Interpolation methods for list: ' + str(a),xlabel='Percentile',ylabel='List item returned',yticks=a)
ax.legend()
plt.show()
```

```python
a = np.arange(4)
a=[1,3,4,5,6,7,8,9]
p = np.linspace(0, 100, 6001)
ax = plt.gca()
lines = [
('linear', None),
('higher', '--'),
('lower', '--'),
('nearest', '-.'),
('midpoint', '-.'),
]
for interpolation, style in lines:
    ax.plot(p, np.percentile(a, p, interpolation=interpolation),label=interpolation, linestyle=style)

ax.set(title='Interpolation methods for list: ' + str(a),xlabel='Percentile',ylabel='List item returned',yticks=a)
ax.legend()
plt.show()
print("numpy percentile is like walking through the number line(here y-axis)")
print("You have to consider points index from 0. Here there are 8 points,index=0 to 7.")
print("10%p means 10% of last index => 10% of 7 .. so index = 0.7")
print("linear method: 0.7 means value[0]+ (value[1]-value[0])*.7")
print(f"{a[0] }+ ({a[1] } -{a[0] }) *.7")
print('-----')
print()
print(np.percentile(a,10,interpolation='linear'))
print(np.percentile(a,10,interpolation='higher'))
print(np.percentile(a,10,interpolation='lower'))
print(np.percentile(a,10,interpolation='nearest'))
print(np.percentile(a,10,interpolation='midpoint'))

print('-----')
print("25%p means 25% of last index => 25% of 7 .. so index = 1.75")
print("linear method: 1.75 means value[1]+ (value[2]-value[1])*.75")
print(f"{a[1] }+ ({a[2] } -{a[1] }) *.75")
print()
print(np.percentile(a,25,interpolation='linear'))
print(np.percentile(a,25,interpolation='higher'))
print(np.percentile(a,25,interpolation='lower'))
print(np.percentile(a,25,interpolation='nearest'))
print(np.percentile(a,25,interpolation='midpoint'))

print()
print("50%p means 50% of last index => 50% of 7 .. so index = 3.5")
print("linear method: 3.75 means value[3]+ (value[4]-value[3])*.5")
print(f"{a[3] }+ ({a[4] } -{a[3] }) *.5")
print(np.percentile(a,50,interpolation='linear'))
print(np.percentile(a,50,interpolation='higher'))
print(np.percentile(a,50,interpolation='lower'))
print(np.percentile(a,50,interpolation='nearest'))
print(np.percentile(a,50,interpolation='midpoint'))


print('----')
print()
print("75%p means 75% of last index => 25% of 7 .. so index = 5.25")
print("linear method: 5.25 means value[5]+ (value[6]-value[5])*.25")
print(f"{a[5] }+ ({a[6] } -{a[5] }) *.25")
print(np.percentile(a,75,interpolation='linear'))
print(np.percentile(a,75,interpolation='higher'))
print(np.percentile(a,75,interpolation='lower'))
print(np.percentile(a,75,interpolation='nearest'))
print(np.percentile(a,75,interpolation='midpoint'))
```
